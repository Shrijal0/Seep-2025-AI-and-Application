```python
#import requierd modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#loading data set
df = sns.load_dataset('titanic')
df.head()

#creating a copy
df_copy = df.copy()

#droping deck and embarked
df_copy.drop(columns = "deck", inplace=True)

df_copy.drop(columns = "embarked", inplace=True)

df_copy.duplicated().sum()

sns.boxplot(df["fare"])
```

```python
q1 = df_copy["fare"].quantile(0.25)
q3 = df_copy["fare"].quantile(0.75)
iqr = q3 - q1
upper = q3 + 1.5 * iqr
lower = q1 - 1.5 * iqr

df_copy = df_copy[(df_copy["fare"] < upper) & (df_copy["fare"] > lower)]

sns.boxplot(df_copy["fare"])
```
