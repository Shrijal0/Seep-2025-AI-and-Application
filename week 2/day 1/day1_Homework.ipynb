```python
import pandas as pd
list = pd.Series([100,200,300,400])
mean = list.mean()
print(mean)
```

```python
list = pd.Series([1,3,5,7,9])
median = list.median()
print(median)
```

```python
list = pd.Series([4,6,8,4,6,8])
mode = list.mode()[0]
print(mode)
```

```python
list = pd.Series([50,75,100,125])
range = list.max() - list.min()
print(range)
```

```python
import matplotlib.pyplot as plt
Dict = pd.DataFrame({
    "Temprature":[22,25,20,28,30]
})
plt.plot(Dict)
plt.xlabel("Days")
plt.ylabel("Temprature")
plt.title("Temprature Graph")
plt.savefig("Temprature.png")
plt.show()
```
